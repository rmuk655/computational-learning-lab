```python
# Your code goes here........ Have fun, call us once you are done:)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
f1 = open("A.csv", "r")
f2 = open("B.csv", "r")
f3 = open("X.csv", "r")
f4 = open("Y.csv", "r")
f5 = open("Z.csv", "r")
A = np.array(pd.read_csv(f1, header = None))
B = np.array(pd.read_csv(f2, header = None))
X = np.array(pd.read_csv(f3, header = None))
Y = np.array(pd.read_csv(f4, header = None))
Z = np.array(pd.read_csv(f5, header = None))
left = 255 - A - 2*B
left = np.array(left)
R = X@Y@Z
R = np.rot90(R, 2)
right = np.flip(R, axis=1)
R = np.hstack((left, right))
plt.imshow(R)
```
