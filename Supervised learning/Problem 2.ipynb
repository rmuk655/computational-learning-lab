```python
class Node:
  def __init__(self, value = 0, children = None, op = None, op_args = None):
    self.value = value
    self.children = children
    self.op = op
    self.op_args = op_args
  def add(self):
    self.value = self.children[0].value + self.children[1].value
  def sub(self):
    self.value = self.children[0].value - self.children[1].value
  def mul(self):
    self.value = self.children[0].value * self.children[1].value
  def div(self):
    self.value = self.children[0].value / self.children[1].value
  def exp(self):
    self.value = self.children[0].value ** self.children[1].value
x = Node(1)
y = Node(2)
b = Node(children = [x, y])
b.add()
print(b.value)
z = Node(3)
a = Node(children = [b, z])
```
