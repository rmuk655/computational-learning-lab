```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/grayscale-v1.png"
im = plt.imread(image_path)
height, width, channels = im.shape
mid_x, mid_y = width // 2, height // 2
q1 = im[:mid_y, :mid_x]
plt.imshow(im)
```

```python
plt.imshow(q1)
```

```python
f = np.fliplr(q1)
g = np.concatenate((q1, f), axis = 1)
h = np.rot90(g, k=2)
i = np.concatenate((g, h), axis = 0)
plt.imshow(i)
```
