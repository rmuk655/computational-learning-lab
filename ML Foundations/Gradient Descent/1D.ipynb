```python
import numpy as np
import matplotlib.pyplot as plt

# Loss function
def f(x):
    return x**2

# Derivative of the loss function
def df(x):
    return 2*x

# Gradient Descent Parameters
x_start = 9  # Starting point/Initial guess
learning_rate = 0.15 # Step size
n_iterations = 10   # Number of iterations

# For storing iteration steps for visualization
x_steps = [x_start]
y_steps = [f(x_start)]

# Values for plotting the parabola
x = np.linspace(-10, 10, 400)
y = f(x)

for i in range(n_iterations):
    plt.figure(figsize=(8, 5))

    # Plot the parabola
    plt.plot(x, y, label='f(x) = x^2')

    # Current position on the parabola
    current_x = x_steps[-1]
    current_y = y_steps[-1]

    # Calculate the gradient
    gradient = df(current_x)

    # Update the position using the gradient descent update rule
    next_x = current_x - learning_rate * gradient  # Update rule
    next_y = f(next_x)

    # Save the next step for visualization
    x_steps.append(next_x)
    y_steps.append(next_y)

    # Plot the current step path
    plt.scatter(x_steps, y_steps, color = 'green')
    plt.plot(x_steps, y_steps, linestyle = '--', marker = 'o', color='blue', label='Gradient Descent Path')

    # Plot the tangent line (derivative at the current point)
    tangent_x = np.linspace(current_x - 1, current_x + 1, 100)
    tangent_y = df(current_x) * (tangent_x - current_x) + current_y
    plt.plot(tangent_x, tangent_y, color='red', linewidth = 0.9, label='Tangent at current step')

    plt.title(f'Gradient Descent Iteration {i+1}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
