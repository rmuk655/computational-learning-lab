```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data points
x = np.array([1, 2, -1, -2])
y = np.array([1, 1, 0, 0])

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Cross entropy loss function
def cross_entropy_loss(x, y, theta, theta_0):
    z = np.dot(x, theta) + theta_0
    y_pred = sigmoid(z)
    return -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))

# Values of theta and theta_0 to use
theta_values = [1, -1, 2, -2]
theta_0_values = [0, 1, -1, 2]

# Plot setup
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Cross Entropy Loss for Different Theta and Theta_0 Values')

for ax, theta, theta_0 in zip(axes.flatten(), theta_values, theta_0_values):
    # Generate range of x values for plotting the sigmoid
    x_range = np.linspace(-3, 3, 100)
    z = theta * x_range + theta_0
    p = sigmoid(z)

    # Plot the sigmoid function
    ax.plot(x_range, p, label=f'Sigmoid: theta={theta}, theta_0={theta_0}')

    # Highlight the data points on the x axes
    ax.scatter(x, np.zeros_like(x), c=['blue' if label == 1 else 'red' for label in y], s = 50)
    ax.vlines(x, 0, sigmoid(theta * x + theta_0), colors='green', linestyles='dotted')

    # Calculate and display the cross entropy loss
    loss = cross_entropy_loss(x, y, theta, theta_0)
    ax.title.set_text(f'Theta: {theta}, Theta_0: {theta_0}, Loss: {loss:.4f}')

    ax.set_ylim(-0.1, 1.1)
    ax.legend()

plt.tight_layout(rect = [0, 0.03, 1, 0.95])
plt.show()
```
