```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_decision_boundary(theta, theta_0, iteration):
    plt.figure(figsize=(8, 6))
    plt.scatter(x[:3, 0], x[:3, 1], color='blue', label='Positive')
    plt.scatter(x[3:, 0], x[3:, 1], color='red', label='Negative')
    x_min, x_max = plt.xlim()
    y_min, y_max = plt.ylim()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    Z = np.dot(theta, np.c_[xx.ravel(), yy.ravel()].T) + theta_0
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, levels=[-np.inf, 0, np.inf], colors=['red', 'blue'], alpha=0.3)
    plt.title(f'Iteration {iteration}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()
```

```python
def Perceptron(x, y, theta, theta_0, n):
    for i in range(n):
        for j in range(len(x)):
            if y[j] * (np.dot(theta, x[j]) + theta_0) <= 0:
                theta += y[j] * x[j]
                theta_0 += y[j]
                break
        plot_decision_boundary(theta, theta_0, i+1)
    return theta, theta_0
```

```python
positive_points = np.array([[-2, 3], [0, 1], [2, -1]])
negative_points = np.array([[-2, 1], [0, -1], [2, -3]])
x = np.vstack((positive_points, negative_points))
y = np.array([1, 1, 1, -1, -1, -1])
theta = np.zeros(2)
theta_0 = 0
n = 10
theta, theta_0 = Perceptron(x, y, theta, theta_0, n)
```

```python
# Plotting the decision boundary
x_values = np.linspace(min(x[:, 0]), max(x[:, 0]), 100)
y_values = -(theta[0] * x_values + theta_0) / theta[1]
plt.plot(x_values, y_values, label='Decision Boundary')
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolors='k', marker='o')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.show()
```
