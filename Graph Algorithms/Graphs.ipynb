```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Define the Data ---
vector_size = np.array([100, 300, 500, 700, 900, 1000])

# Data for integer and float times
time_1 = np.array([0.000002519, 0.000019342, 0.000050907, 0.000099165, 0.000168237, 0.000212641])
time_2 = np.array([0.000002428, 0.000018479, 0.000050405, 0.000097834, 0.000158997, 0.000205710])

# Calculate Speedup
speedup = (time_1 - time_2) / time_1 * 100

# --- 2. Create the Integrated Line Graph ---
plt.figure(figsize=(10, 6))

plt.plot(vector_size, time_1, marker='o', linestyle='-', color='red', label='Vector size vs time for integer', linewidth=2)
plt.plot(vector_size, time_2, marker='s', linestyle='--', color='blue', label='Vector size vs time for float', linewidth=2)

plt.title('Execution time Vs Vector size for integers and float comparison', fontsize=16)
plt.xlabel('Vector size', fontsize=12)
plt.ylabel('Time', fontsize=12)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

# --- 3. Create the Speedup Bar Graph ---
plt.figure(figsize=(10, 6))
plt.bar(vector_size.astype(str), speedup, color='lightgreen', edgecolor='black')
plt.xlabel('Vector Size', fontsize=12)
plt.ylabel('Speedup (%)', fontsize=12)
plt.title('Speedup vs Vector Size', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Define the Data ---
int_n = np.array([5, 10, 15, 20])

# Data for the first line (Website 1 Views)
time1 = np.array([0.000000038, 0.000000040, 0.000000043, 0.000000052])

# Data for the second line (Website 2 Views)
time2 = np.array([0.000000036, 0.000000036, 0.000000037, 0.000000038])

speedup = (time1-time2)/time1*100

# --- 2. Create the Integrated Line Graph ---
def create_dual_line_graph():
    """Generates a single graph with two distinct lines."""

    # Set the size of the plot
    plt.figure(figsize=(10, 6))

    # Plot the first line
    plt.plot(
        int_n,
        time_1,
        marker='o',             # Circle markers
        linestyle='-',
        color='red',            # Red line
        label='Time for integer',# Label for the legend
        linewidth=2
    )

    # Plot the second line
    plt.plot(
        int_n,
        time_2,
        marker='s',             # Square markers
        linestyle='--',         # Dashed line
        color='blue',           # Blue line
        label='Time for float',# Label for the legend
        linewidth=2
    )

    # Add title and labels
    plt.title('Execution Time vs Number n factorial', fontsize=16)
    plt.xlabel('Number n', fontsize=12)
    plt.ylabel('Execution Time', fontsize=12)

    # Show the legend to differentiate the lines
    plt.legend(loc='upper left', fontsize=10)

    # Add a grid for better readability
    plt.grid(True, linestyle=':', alpha=0.6)

    # Rotate x-axis labels if necessary (good for longer labels)
    plt.xticks(rotation=0)

    # Display the plot
    plt.show()

if __name__ == "__main__":
    create_dual_line_graph()
```
